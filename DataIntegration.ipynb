```
import pandas as pd
```

```
df = pd.read_csv('/content/COVID_county_data.csv.gz')
df['county']=df['county']+' County'
df.head(10)
```

```
df1 = pd.read_csv('/content/acs2017_census_tract_data.csv.gz')


```

```
#df1.columns
acs_data = df1[['State', 'County','TotalPop','Poverty','IncomePerCap']]
acs_data.head(10)
acs_data['Poverty(In Num)'] = (acs_data['TotalPop'] * acs_data['Poverty'])/ 100
acs_data['TotalIncome'] = (acs_data['TotalPop'] * acs_data['IncomePerCap'])
acs_data.head(10)
```

```
acs_data.drop(columns = ['Poverty', 'IncomePerCap'],inplace=True)
acs_data.head(5)
```

```
acs_agg = acs_data.groupby(['State','County']).max()
acs_agg.head(5)

acs_agg['Poverty']= (acs_agg['Poverty(In Num)']/acs_agg['TotalPop'])*100
acs_agg['IncomePerCap']= acs_agg['TotalIncome']/acs_agg['TotalPop']

acs_agg.drop(columns = ['TotalIncome'],inplace=True)

acs_agg.head(5)
```

```
#Loudoun County Virginia, Washington County Oregon, Harlan County Kentucky, Malheur County Oregon

result = acs_agg.loc[[('Oregon', 'Washington County'),('Virginia','Loudoun County'),('Kentucky','Harlan County'),('Oregon','Malheur County')]]

result

```

```
#TotalCases - total number of COVID cases for this county as of February 20, 2021
#Dec2020Cases - number of COVID cases recorded in this county in December of 2020
#df.columns.rename(['State', 'County'], inplace =True)
df.head(5)

df.rename(columns={"state": "State", "county": "County"},inplace=True)
covid_agg = df.groupby(['State','County']).max()
covid = covid_agg.rename(columns={"cases": "total cases", "deaths": "total deaths"})
covid.drop(columns = ['fips'],inplace=True)
result_covid = covid.loc[[('Oregon', 'Washington County'),('Virginia','Loudoun County'),('Kentucky','Harlan County'),('Oregon','Malheur County')]]
result_covid


```

```
#TotalDeaths - total number of COVID deaths for this county as of February 20, 2021
#Dec2020Deaths - number of COVID deaths recorded in this county in December of 2020
covid_dec=df[(df['date'] > '2020-11-30') & (df['date'] < '2021-01-01')]
covid_dec.rename(columns={"cases": "dec_cases", "deaths": "dec_deaths"},inplace=True)
covid_dec_agg = covid_dec.groupby(['State','County']).max()
covid_dec_agg.drop(columns = ['fips'],inplace=True)
result_covid1 = covid_dec_agg.loc[[('Oregon', 'Washington County'),('Virginia','Loudoun County'),('Kentucky','Harlan County'),('Oregon','Malheur County')]]
result_covid1
```

```
r = pd.concat([covid_dec_agg,covid], axis=1)
#result_covid2 = r.loc[[('Oregon')]]
result_covid3 = pd.concat([acs_agg,r], axis = 1)
final_result = result_covid3.loc[['Oregon']]
result_covid3=result_covid3.groupby(['State']).sum()
result_covid3
```

```
R = final_result['total cases'].corr(final_result['Poverty'])
print(R)
final_result['CovidTotalCases'] = (final_result['total cases'] * 100000 / final_result['TotalPop'])
final_result
final_result['CovidDeathCases'] = (final_result['total deaths'] * 100000 / final_result['TotalPop'])
final_result
final_result['CovidDecTotalCases'] = (final_result['dec_cases'] * 100000 / final_result['TotalPop'])
final_result
final_result['CovidDecDeathCases'] = (final_result['dec_deaths'] * 100000 / final_result['TotalPop'])
final_result
# Entire USA result
U = result_covid3['total cases'].corr(result_covid3['Poverty'])
print(U)
result_covid3['CovidTotalCases'] = (result_covid3['total cases'] )

result_covid3['CovidDeathCases'] = (result_covid3['total deaths'] )

result_covid3['CovidDecTotalCases'] = (result_covid3['dec_cases'])

result_covid3['CovidDecDeathCases'] = (result_covid3['dec_deaths'])
result_covid3
```

```
#COVID total cases vs. % population in poverty
R1 = result_covid3['CovidTotalCases'].corr(result_covid3['Poverty'])
print(R1)
#COVID total deaths vs. % population in poverty
R2 = result_covid3['CovidDeathCases'].corr(result_covid3['Poverty'])
print(R2)
#COVID total cases vs. Per Capita Income level
R3 = result_covid3['CovidTotalCases'].corr(result_covid3['IncomePerCap'])
print(R3)
#COVID total deaths vs. Per Capita Income level
R4 = result_covid3['CovidDeathCases'].corr(result_covid3['IncomePerCap'])
print(R4)
#COVID cases during December 2020 vs. % population in poverty
R5 = result_covid3['CovidDecTotalCases'].corr(result_covid3['Poverty'])
print(R5)
#COVID deaths during December 2020 vs. % population in poverty
R6 = result_covid3['CovidDecDeathCases'].corr(result_covid3['Poverty'])
print(R6)
#COVID cases during December 2020 vs. Per Capita Income level
R7 = result_covid3['CovidDecTotalCases'].corr(result_covid3['IncomePerCap'])
print(R7)
#COVID cases during December 2020 vs. Per Capita Income level
R8 = result_covid3['CovidDecDeathCases'].corr(result_covid3['IncomePerCap'])
print(R8)
```
